```python
import os
import pandas as pd
import itertools
import csv
from scipy import stats
import numpy as np

def metabolite(file_name):
    #Read data
    filtered_data = pd.read_csv(file_name, header=None, na_values=["", "NaN"], keep_default_na=True)
    
    # Get metabolite names. Skip the first 2 rows (headers + Sample IDs)
    names = filtered_data.iloc[2:, 0].tolist()
        
    # Generate all 2-element combinations
    pairs = list(itertools.combinations(names, 2))
    results = []

    temp = 0 #DELETE THIS LATER --------------------------------------------------------------------
    for met1, met2 in pairs:
        #if (temp < 1): #DELETE THIS LATER --------------------------------------------------------------------
            
            # Extract rows with values (column 2+)
            l1 = filtered_data[filtered_data.iloc[:, 0] == met1].iloc[:, 1:]
            l2 = filtered_data[filtered_data.iloc[:, 0] == met2].iloc[:, 1:]
    
            #Convert values to float
            row1 = l1.astype(float).values.flatten()
            row2 = l2.astype(float).values.flatten()
            #print("Row 1:", row1)
            #print("Row 2:", row2)
            #temp = temp + 1
    
            # Create a mask for positions where BOTH are non-NaN
            mask = ~np.isnan(row1) & ~np.isnan(row2)
            #print("Mask:",mask)
            # Apply mask to array
            r1_masked = row1[mask]
            #print(r1_masked)
            r2_masked = row2[mask]
            #print(r2_masked)

    
    #       If less than three paired points: let corr and pval be NaN. Else: calculate them
            if len(r1_masked) < 3:
                corr, pval = np.nan, np.nan
            else:
                # Calculate Spearman correlation
                corr, pval = stats.spearmanr(r1_masked, r2_masked)
            temp += 1
            
            # Append metabolite names + correlation + p-value
            results.append((met1, met2, corr, pval))


    
    
    # Create output directory + output file path
    parts = os.path.basename(file_name).split("_")
    prefix = "_".join(parts[:2])
    os.makedirs("6. Metabolite Pairs", exist_ok=True)
    output_file = os.path.join("6. Metabolite Pairs", f"COMBINATIONS_{prefix}.csv")
    
    with open(output_file, "w", newline='', encoding="utf-8") as outfile:
        writer = csv.writer(outfile)
        writer.writerow(["Metabolite 1", "Metabolite 2", "Spearman Coefficient", "p-value"])
        writer.writerows(results)

    #print(f"Saved {output_file} with {len(results)} metabolite pairs.")

# ------------- Run -------------
metabolite("./5. Normalized/pls_VECPAC_CTRL_normalized.csv")
metabolite("./5. Normalized/pls_LPS_CTRL_normalized.csv")
metabolite("./5. Normalized/pls_DSS_CTRL_normalized.csv")
```
